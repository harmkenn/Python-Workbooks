import one column from an excel spreadsheet

```python
import pandas as pd

def import_excel_column(file_path, column_name):
  """
  Imports one column from an Excel spreadsheet.

  Args:
    file_path: The path to the Excel spreadsheet.
    column_name: The name of the column to import.

  Returns:
    A list of the values in the specified column.
  """

  # Read the Excel spreadsheet into a pandas DataFrame.
  df = pd.read_excel(file_path)

  # Get the values in the specified column.
  column_values = df[column_name].tolist()

  # Return the list of values.
  return column_values

# Import the column "Name" from the Excel spreadsheet "example.xlsx".
column_values = import_excel_column("symbols.xlsx", "Symbol")

# Print the column values.
print(column_values)

# Turn that list into a pandas dataframe

nf2021 = pd.DataFrame(column_values)
```

```python
import datetime

def find_closest_wednesday(year):
  """
  Finds the Wednesday closest to 25 July for the specified year.

  Args:
    year: The year to find the Wednesday closest to 25 July for.

  Returns:
    The date of the Wednesday closest to 25 July for the specified year.
  """

  # Get the date of 25 July for the specified year.
  date = datetime.date(year, 7, 25)

  # Find the next Wednesday after 25 July.
  next_wednesday = date + datetime.timedelta(days=7 - date.weekday())

  return next_wednesday

# Find the Wednesday closest to 25 July for the past 10 years.
for year in range(2013, 2023):
  date = find_closest_wednesday(year)
  date1 = date + datetime.timedelta(days=1)


  print(f"The Wednesday closest to 25 July in {year} is {date} and {date1}")
```
