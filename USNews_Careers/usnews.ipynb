```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://money.usnews.com/careers/best-jobs/search"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Now you can work with the 'soup' object to extract the information you need
    # For example, you can find elements by their class, id, tag, etc.
    # soup.find(), soup.find_all(), soup.select() are useful methods
    
    # For demonstration, let's print the title of the webpage
    print("Title of the webpage:", soup.title.text)
    
else:
    print("Failed to retrieve webpage")
```
