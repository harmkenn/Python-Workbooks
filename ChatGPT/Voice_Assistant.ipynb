```python
import speech_recognition as sr
import nltk
import pyttsx3

# Initialize the speech recognition engine
r = sr.Recognizer()

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Define a function to process the user's input and generate a response
def process_input(input_text):
    # Use NLTK to tokenize the input text
    tokens = nltk.word_tokenize(input_text)
    
    # Define the chatbot's response based on the user's input
    if 'hello' in tokens:
        response = 'Hi there!'
    elif 'goodbye' in tokens:
        response = 'Goodbye!'
    else:
        response = 'Sorry, I did not understand that.'
        
    # Use the text-to-speech engine to generate a spoken response
    engine.say(response)
    engine.runAndWait()

# Define a function to listen for the user's input and pass it to the processing function
def listen():
    with sr.Microphone() as source:
        print('Say something...')
        audio = r.listen(source)
        
    try:
        # Use the speech recognition engine to convert the audio to text
        input_text = r.recognize_google(audio)
        print('You said: ' + input_text)
        
        # Process the user's input
        process_input(input_text)
        
    except sr.UnknownValueError:
        print('Sorry, I could not understand what you said.')
    except sr.RequestError as e:
        print('Could not request results from speech recognition service; {0}'.format(e))

# Continuously listen for user input
while True:
    listen()
```
