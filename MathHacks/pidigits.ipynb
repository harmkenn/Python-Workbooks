```python
import math

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def estimate_pi(num_terms):
    result = 0
    k = 0
    while True:
        term = (factorial(4 * k) * (1103 + 26390 * k)) / ((factorial(k) ** 4) * (396 ** (4 * k)))
        result += term
        if abs(term) < 1e-15:
            break
        k += 1

    return 1 / (result * (2 * math.sqrt(2) / 9801))

# Calculate π using Ramanujan series with 10 terms
num_terms = 99
approximated_pi = estimate_pi(num_terms)
print(f"Approximation of π using {num_terms} terms: {approximated_pi:.99f}")
print(f"Actual π value:                    {math.pi:.99f}")
```
