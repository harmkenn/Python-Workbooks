```python
import xml.etree.ElementTree as ET
import re

# Parse the KML file
tree = ET.parse('AL24 JOA 05.kml')
root = tree.getroot()

# Find and extract coordinates for each path
all_paths = []
for placemark in root.findall('.//{http://www.opengis.net/kml/2.2}Placemark'):
    path_coordinates = []
    for coord in placemark.findall('.//{http://www.opengis.net/kml/2.2}coordinates'):
        path_coordinates.extend(re.findall(r'(-?\d+\.\d+),(-?\d+\.\d+),0', coord.text))
    all_paths.append(path_coordinates)

# Now all_paths is a list containing coordinates for each path
for i, path in enumerate(all_paths):
    print(f"Path {i+1} coordinates:")
    print(path)
```
